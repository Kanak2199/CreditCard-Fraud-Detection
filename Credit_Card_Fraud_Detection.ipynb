```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('creditcard.csv')
df.head()
```

```python
df['Class'].value_counts()
```

```python
df.hist(bins=30, figsize=(25, 25))
```

```python
df.describe()
```

```python
from sklearn.preprocessing import RobustScaler
n_df = df.copy()
n_df['Amount'] = RobustScaler().fit_transform(n_df['Amount'].to_numpy().reshape(-1, 1))
```

```python
time = n_df['Time']
n_df['Time'] = (time - time.min()) / (time.max() - time.min())
n_df
```

```python
n_df = n_df.sample(frac=1, random_state=1)
n_df
```

```python
train, test, val = n_df[:200000], n_df[200000:245000], n_df[245000:]
train['Class'].value_counts(), test['Class'].value_counts(), val['Class'].value_counts()
```

## **Supervised Learning On Imbalance Dataset**

```python
train_np, test_np, val_np = train.to_numpy(), test.to_numpy(), val.to_numpy()
train_np.shape, test_np.shape, val_np.shape
```

```python
x_train, y_train = train_np[:, :-1], train_np[:, -1]
x_test, y_test = test_np[:, :-1], test_np[:, -1]
x_val, y_val = val_np[:, :-1], val_np[:, -1]
x_train.shape, y_train.shape, x_test.shape, y_test.shape, x_val.shape, y_val.shape
```

```python
from sklearn.linear_model import LogisticRegression
logistic_model = LogisticRegression()
logistic_model.fit(x_train, y_train)
logistic_model.score(x_train, y_train)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_val, logistic_model.predict(x_val), target_names=['Not Fraud', 'Fraud']))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(max_depth=2, n_jobs=-1)
rf.fit(x_train, y_train)
y_pred = rf.predict(x_val)
print(classification_report(y_val, y_pred, target_names=['Not Fraud', 'Fraud']))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
gbc = GradientBoostingClassifier(n_estimators=50, learning_rate=1.0, max_depth=1, random_state=0)
gbc.fit(x_train, y_train)
y_pred = gbc.predict(x_val)
print(classification_report(y_val, y_pred , target_names=['Not Fraud', 'Fraud']))
```

```python
from sklearn.svm import LinearSVC
svc = LinearSVC(class_weight='balanced')
svc.fit(x_train, y_train)
y_pred = svc.predict(x_val)
print(classification_report(y_val, y_pred, target_names=['Not Fraud', 'Fraud']))
```

## **Neural Networks on Imbalanced Dataset**

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import InputLayer, Dense, BatchNormalization
from tensorflow.keras.callbacks import ModelCheckpoint

shallow_nn = Sequential()
shallow_nn.add(InputLayer((x_train.shape[1],)))
shallow_nn.add(Dense(2, 'relu'))
shallow_nn.add(BatchNormalization())
shallow_nn.add(Dense(1, 'sigmoid'))

checkpoint = ModelCheckpoint('shallow_nn.keras', save_best_only=True)
shallow_nn.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
shallow_nn.summary()
```

```python
shallow_nn.fit(x_train, y_train, validation_data=(x_val, y_val), epochs=5, callbacks=checkpoint)
```

```python
def neural_net_predictions(model, x):
  return (model.predict(x).flatten() > 0.5).astype(int)
neural_net_predictions(shallow_nn, x_val)
```

```python
print(classification_report(y_val, neural_net_predictions(shallow_nn, x_val), target_names=['Not Fraud', 'Fraud']))
```

# **Unsupervised Learning - Anamoly Detection**

## **Isolation Forest**

```python
from sklearn.ensemble import IsolationForest

X = n_df.drop(columns=['Class'])
iso_forest = IsolationForest(contamination=0.2, random_state=42)
iso_forest.fit(X)

# Predict anomalies (-1 for outliers, 1 for normal data points)
n_df['anomaly'] = iso_forest.predict(X)
```

```python
# Plotting the first two features (just for visualization, you can plot different or more features)
plt.figure(figsize=(10, 6))
plt.scatter(n_df['V1'], n_df['V2'], c=n_df['anomaly'], cmap='coolwarm', edgecolors='k', s=20)
plt.title("Anomaly Detection Using Isolation Forest")
plt.xlabel("V1")
plt.ylabel("V2")
plt.colorbar(label='Anomaly (-1: Anomaly, 1: Normal)')
plt.show()
```

```python
# Show the number of anomalies detected
print("Number of anomalies detected:", n_df['anomaly'].value_counts()[-1])
```

```python
# Filtering the anomalies
anomalies = n_df[n_df['anomaly'] == -1]
print("Anomalies detected:\n", anomalies)
```

 ## **Local Outlier Factor**

```python
from sklearn.neighbors import LocalOutlierFactor

X = n_df.drop(columns=['Class'])
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.05)

n_df['anomaly'] = lof.fit_predict(X)

# The LOF algorithm returns 1 for normal data points and -1 for anomalies
# Invert the prediction to match the typical anomaly label (-1 for outliers, 1 for normal data points)
n_df['anomaly'] = n_df['anomaly'].apply(lambda x: 1 if x == 1 else -1)
```

```python
# Visualize anomalies (using the first two features for demonstration purposes)
plt.figure(figsize=(10, 6))
plt.scatter(n_df['V1'], n_df['V2'], c=n_df['anomaly'], cmap='coolwarm', edgecolors='k', s=20)
plt.title("Anomaly Detection Using Local Outlier Factor (LOF)")
plt.xlabel("V1")
plt.ylabel("V2")
plt.colorbar(label='Anomaly (-1: Anomaly, 1: Normal)')
plt.show()
```

```python
# Show the number of anomalies detected
print("Number of anomalies detected:", n_df['anomaly'].value_counts()[-1])
```

```python
# Optionally, filter and display anomalies
anomalies = n_df[n_df['anomaly'] == -1]
print("Anomalies detected:\n", anomalies)
```

## **DBSCAN for Anomaly Detection**

```python
# Import necessary libraries
from sklearn.cluster import DBSCAN

X = n_df.drop(columns=['Class'])
dbscan = DBSCAN(eps=0.5, min_samples=5)  # eps is the maximum distance between points in a cluster
n_df['anomaly'] = dbscan.fit_predict(X)
n_df['anomaly'] = n_df['anomaly'].apply(lambda x: 1 if x != -1 else -1)
```

```python
# Visualize anomalies (using the first two features for demonstration purposes)
plt.figure(figsize=(10, 6))
plt.scatter(n_df['V1'], n_df['V2'], c=n_df['anomaly'], cmap='coolwarm', edgecolors='k', s=20)
plt.title("Anomaly Detection Using DBSCAN")
plt.xlabel("V1")
plt.ylabel("V2")
plt.colorbar(label='Anomaly (-1: Anomaly, 1: Normal)')
plt.show()
```

```python
# Show the number of anomalies detected
print("Number of anomalies detected:", n_df['anomaly'].value_counts()[-1])
```

```python
# Optionally, filter and display anomalies
anomalies = n_df[n_df['anomaly'] == -1]
print("Anomalies detected:\n", anomalies)
```
